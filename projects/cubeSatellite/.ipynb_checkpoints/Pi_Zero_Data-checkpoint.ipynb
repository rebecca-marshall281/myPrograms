# Raspberry Pi Zero Testing for the BSS Two Unit Cub Satillite 

This notebook was built using the data collected in the testing of a Raspberry Pi Zero. The intent of the data is to show that a Raspberry Pi Zero can operate in several environments: static, near near vacuum, hot  temperature (XX degress Celsius), cold temperature (-XX degrees Celsius) and high impact while preforming a simple computational program.

For each test the Raspberry Pi Zero was booted, placed in the environment and then a program was executed that had the Raspberry Pi Zero compute 250 000 calculations using the equation x to the power of two. The resulting data was then graphed to look for any missed calculations that could be an indication of a change in the performance of the Pi Zero due to the environment. This program was intended to run approximately 10 minutes which will potentially be the longest amount of time the Pi will need to survive each environment. The resulting data was then graphed to look for any missed calculations that could be an indication of the Model Zero enduring a change in performance because of the environment it was exposed to. The data was graphed to determine if there were any anomalies (miscalculations) in the data collected. The time taken to complete the calculations is also an indicator for a change in the Pi's performance.

```python
#import of libraries.
#this cell must be run each time the notebook is started.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
```

## Static Test:
The first test the Raspberry Pi Zero under went was a static test in which it was operated at room temperature (21 degress Celsius).

```python
Image("Images/PIZSTAT.jpg")
```

```python
#this cell imports the csv data file and displays its information to ensure all data has been included.
data = pd.read_csv('Raw_Data/static-pyzero_data1.csv')
data.info()
```

```python
#this cell plots the data as the reult as a function of the iteration
x_data = data['Iteration']
y_data = data['Result']
plt.plot(x_data, y_data)
plt.show()
```

As can be seen from the graph the Raspberry Pi Zero did not have any troubles running through the calculations in the static test.

## Near Vacuum:
This test was done using the school's bell jar under which the Raspberry Pi Zero was placed. A ring of plumbers putty between the jar and its base allowed for the running of the cables. 

```python
Image("Images/Model3Vac.jpg")
```

```python
#this cell imports the csv data file and displays its information to ensure all data has been included.
data = pd.read_csv('Raw_Data/pressure-pizero_630.48.csv')
data.info()
```

```python
#this cell plots the data as the reult as a function of the iteration
x_data = data['Iteration']
y_data = data['Result']
plt.plot(x_data, y_data)
plt.xlabel('Iteration')
plt.ylabel('Result')
plt.title('Near Vacuum Data')
plt.show()
```

As can be seen from the graph the Raspberry Pi Zero did not have any troubles running through the calculations in the static test.
