# Raspberry Pi Model B Testing for the BSS Two Unit Cub Satillite 

This notebook was built using the data collected in the testing of a Raspberry Pi 3 Model B. The intent of the data is to show that a Raspberry Pi Model B can operate is several environments: static, near near vacuum, hot  temperature (XX degress Calculus), cold temperature (-XX degrees Calculus) and high impact while preforming a task.

For each test the Model 3 was booted, placed in the environment and then a program was exicuted that had the Model 3 compute 2.2 million calculations using the equation x to the power of two. The resulting data was then graphed to look for any missed calculations that could be an indication of the Model 3 in the environment.

```python
#import of libraries.
#this cell must be run each time the notebook is started.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
```

## Static Test:
The first test the Model 3 under went was a static test in which it was operated at room temperature (21 degress Calculus)

```python
Image("Images/Model3Stat.jpg")
```

```python
#this cell imports the csv data file and displays its information to ensure all data has been included.
data = pd.read_csv('Raw_Data/static-pimodelb-data1.csv')
data.info()
```

```python
#this cell plots the data as the reult as a function of the iteration
x_data = data['Iteration']
y_data = data['Result']
plt.plot(x_data, y_data)
plt.show()
```

As can be seen from the graph the Model 3 did not have any troubles running through the calculations in the static test.
